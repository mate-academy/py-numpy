````python
# Python code
def unknown_signature(lst, mask, new_value):
    for i in range(len(lst)):
        if mask[i]:
            lst[i] = new_value

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0
%timeit unknown_signature(lst, mask, new_value)
lst
````

````python
def replace_list(lst, mask, new_value):
    lst = np.array(lst, dtype=np.int8)
    lst[mask] = new_value
    return lst

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0

%timeit replace_list(lst, mask, new_value)
replace_list(lst, mask, new_value)
````

### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
%timeit unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
def change_lists(lst1, lst2):
    lst1 = np.array(lst1, dtype=np.int8)
    lst2 = np.array(lst2, dtype=np.int8)
    
    bitwise_complement = ~lst1
    bitwise_and = lst1 & lst2
    bitwise_or = lst1 | lst2
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]

%timeit change_lists(list1, list2)
change_lists(list1, list2)
````

### Task 7.3 Optional

````python
# Python code
def unknown_signature(lst):
    result = []
    for num in lst:
        if num % 2 == 0 and (num % 3 == 0 or num > 10):
            modified_num = (num ^ 7) + 2
            result.append(modified_num)
    return result

my_list = [6, 9, 12, 14, 17, 20]
%timeit unknown_signature(my_list)
unknown_signature(my_list)
````

````python
def list_mat_operations(lst):
    lst = np.array(lst, dtype=np.int8)
    return [(num ^ 7) + 2 for num in lst if num % 2 == 0 & (num % 3 == 0 | num > 10)]

my_list = [6, 9, 12, 14, 17, 20]
%timeit list_mat_operations(my_list)
list_mat_operations(my_list)
````
