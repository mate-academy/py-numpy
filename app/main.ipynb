````python
# Python code
def unknown_signature(lst, mask, new_value):
    for i in range(len(lst)):
        if mask[i]:
            lst[i] = new_value

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0
unknown_signature(lst, mask, new_value)
lst
````

````python
def unknown_signature_np(lst, mask, new_value):
    lst_np = np.array(lst)
    mask_np = np.array(mask)
    lst_np[mask_np] = new_value
    return lst_np.tolist()

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0

result_np = unknown_signature_np(lst, mask, new_value)
result_np
````

### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
def unknown_signature_np(lst1, lst2):
    bitwise_complement = np.invert(lst1)
    bitwise_and = np.bitwise_and(lst1, lst2)
    bitwise_or = np.bitwise_or(lst1, lst2)
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]

result_bitwise_complement_np, result_bitwise_and_np, result_bitwise_or_np = unknown_signature_np(np.array(list1), np.array(list2))
result_bitwise_complement_np, result_bitwise_and_np, result_bitwise_or_np
````

### Task 7.3 Optional

````python
# Python code
def unknown_signature(lst):
    result = []
    for num in lst:
        if num % 2 == 0 and (num % 3 == 0 or num > 10):
            modified_num = (num ^ 7) + 2
            result.append(modified_num)
    return result

my_list = [6, 9, 12, 14, 17, 20]
result_python_style = unknown_signature(my_list)
result_python_style
````

````python
def unknown_signature_np(lst):
    lst_np = np.array(lst)
    mask = (lst_np % 2 == 0) & ((lst_np % 3 == 0) | (lst_np > 10))
    modified_num = (lst_np[mask] ^ 7) + 2
    return modified_num.tolist()

my_list = [6, 9, 12, 14, 17, 20]
result_numpy_style = unknown_signature_np(my_list)
result_numpy_style
````
